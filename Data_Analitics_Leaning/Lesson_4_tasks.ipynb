```python
# 4.9
# Create a wait_time column that will store the difference between the arrival time (arrived_at)
# and the order time (start_at) in minutes.

# To convert the result to minutes, 
# you can use the .astype('timedelta64[m]') method.
```

```python
taxi = pd.read_csv('~//OneDrive//Karpov_courses//Data_Analist//Lessons_codes//Lesson_3//3_taxi_peru.csv', sep = ';')
taxi['start_at'] = pd.to_datetime(taxi['start_at'])
taxi['end_at'] = pd.to_datetime(taxi['end_at'])
taxi['arrived_at'] = pd.to_datetime(taxi['arrived_at'])
taxi
```

```python
taxi['wait_time'] = (taxi['arrived_at'] - taxi['start_at']).astype('timedelta64[m]')
```

```python
# 4.10
# Let's see which drivers were late for orders by a certain time (start_type == 'reserved').
# Lateness is considered waiting time wait_time > 0.0.

# Group the filtered data by driver_id and as an answer specify the id of the driver 
# who was late for the order of the type we are interested in the most number of times
(
    taxi
    .query("start_type == 'reserved'")
    .query("wait_time > 0.0")
    .groupby(['driver_id'], as_index = False)
    .agg({'journey_id' : 'count'})
    .rename(columns = {'journey_id' : 'number_of_delayed_journeys'})
    .sort_values('number_of_delayed_journeys', ascending = False)
)
# answer = 406921adcca37705ef527b4246c0cfea
```

```python
# 4.11
# Add a month column to store the number of the month in which the order was placed (start_at)
# and a weekday column to store the name of the day of the week.

taxi['month'] = taxi.start_at.dt.month
taxi['weekday'] = taxi['start_at'].dt.strftime('%A')
```

```python
# 4.15
# We can also look at a few simple metrics showing the number of unique users over a given period:

# DAU (daily active users) - the number of unique users per day
# WAU (weekly active users) - the number of unique users per week
# MAU (monthly active users) - the number of unique users per month
# Active users are those who used the application and made at least one order for the selected period of time.
# Plot the MAU and choose the correct statements.

unique_users_by_month = (
    taxi
    .groupby(['month'], as_index = False)
    .agg({'user_id' : 'nunique'})
    .rename(columns = {'user_id' : 'number_of_unique_users'})
)


unique_users_by_month.plot.line(x = 'month', y = 'number_of_unique_users')
```
