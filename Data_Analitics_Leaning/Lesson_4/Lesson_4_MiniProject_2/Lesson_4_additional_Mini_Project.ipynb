```python
import pandas as pd
from pathlib import Path
```

```python
# And a small mini-project on data filtering!

# There is a table with company records, including their TINs, 
# and a separate text file with a set of TINs (head_inn column) stored in folder 4_inn.

# inn.xls - table in Excel format
# necessary_inn.txt - list of required TINs in text format

# Tasks:
# to extract from the table records with TINs given in the text file
# write the result to file selected_inn.csv
```

```python
path_to_df_with_all_INNs = Path(Path.cwd(), '4_inn.xls')
df_with_all_INNs = pd.read_excel(path_to_df_with_all_INNs)
df_with_all_INNs = df_with_all_INNs.drop(columns = ['Unnamed: 0', '#'])
```

```python
path_to_neccesary_INNs = Path(Path.cwd(), '4_necessary_inn.txt')
neccesary_INNs = (
    pd.read_csv(path_to_neccesary_INNs, header = None)
    .rename(columns={0:'head_inn'})
)
```

```python
df_with_nccsry_INNs = df_with_all_INNs.merge(neccesary_INNs, on = 'head_inn')
df_with_nccsry_INNs
```

```python
path_to_selected_inn_csv = Path(Path.cwd(), 'selected_inn.csv')

df_with_nccsry_INNs.to_csv(path_to_selected_inn_csv)
```
